```python
import pandas as pd
from geopy.distance import geodesic
from sklearn.neighbors import NearestNeighbors

# Read the CSV files
abCoffee_outlets = pd.read_csv('abCoffee_outlets.csv')
cafe_data_2 = pd.read_csv('cafe_data_2.csv')

# Define weights for different outlet types
weights = {'Coffee': 10, 'Tea': 5, 'Cafe': 15, 'Other': 8}

# Function to calculate distance between two coordinates
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).meters

# Function to create clusters and calculate metrics
def create_clusters(outlet_type):
    clusters = []
    filtered_data = cafe_data_2[cafe_data_2['type'] == outlet_type]

    for _, outlet in filtered_data.iterrows():
        outlet_location = (outlet['Latitude'], outlet['Longitude'])
        cluster = {'outlet_location': outlet_location, 'coordinates': [], 'metric': 0}

        for _, cafe in filtered_data.iterrows():
            cafe_location = (cafe['Latitude'], cafe['Longitude'])
            distance = calculate_distance(outlet_location, cafe_location)

            if distance <= 100:
                cluster['coordinates'].append(cafe_location)
                cluster['metric'] += distance * weights.get(cafe['type'], 1)

        clusters.append(cluster)
    return clusters

# Create clusters for different outlet types
outlet_clusters = {}
for outlet_type in ['Coffee', 'Tea', 'Cafe', 'Other']:
    outlet_clusters[outlet_type] = create_clusters(outlet_type)

# Find the cluster with the highest metric for abCoffee outlets
max_metric_cluster = max(outlet_clusters['Coffee'], key=lambda x: x['metric'])

# Extract coordinates of the selected cluster
selected_cluster_coordinates = max_metric_cluster['coordinates']



```
